```python
import random

class ListNode:
    def __init__(self, value1, next1=None):
        self.value = value1
        self.next = next1


class List:
    def __init__(self):
        self._first = None
        self._last = None
        self.ln = 0

    def __len__(self):
        return self.ln

    def front(self): #  == my_array[0]
        return self._first.value

    def last(self): #  == my_array[-1]
        return self._last.value

    def empty(self):
        return self.ln == 0

    def push_front(self, value):
        self._first = ListNode(value, self._first)
        if self._last is None:
            self._last = self._first
        self.ln += 1

    def push_back(self, value):
        toadd = ListNode(value)
        if self.empty():
            self._first = toadd
            self._last = toadd
        else:
            self._last.next = toadd
            self._last = toadd

        self.ln += 1

    def pop_front(self):
        if self.ln == 0:
            print("Can not pop from empty list")
            return
        if self.ln == 1:
            self._first = None
            self._last = None
            self.ln = 0
            return
        self._first = self._first.next
        self.ln -= 1

    def pop_back(self):
        if self.ln == 0:
            print("Pop from empty list")
            return
        if self.ln == 1:
            self._first = None
            self._last = None
            self.ln = 0
            return
        cur = self._first
        for i in range(self.ln - 2):
            cur = cur.next
        cur.next = None
        self._last = cur
        self.ln -= 1

    def clear(self):
        self._first = None
        self._last = None
        self.ln = 0

    def print(self):
        cur = self._first
        for i in range(self.ln):
            print(cur.value, end=" ")
            cur = cur.next
        print()

    def pop_all_value(self, value1):
        cur = self._first
        while cur.value == value1:
            self._first = cur.next
            self.ln -= 1
            cur = cur.next
        for i in range(self.ln-1):
            if cur.next.value == value1:
                cur.next = cur.next.next
                self.ln -= 1
            else:
                cur = cur.next
        self.set_correct_last()



    def pop_one_value(self, value1):
        cur = self._first
        while cur.value == value1:
            self._first = cur.next
            self.ln -= 1
            cur = cur.next
            self.set_correct_last()
            return
        for i in range(self.ln-1):
            if cur.next.value == value1:
                cur.next = cur.next.next
                self.ln -= 1
                self.set_correct_last()
                return
            else:
                cur = cur.next


    def set_correct_last(self):
        if self._first == None:
            self._last = None
            return
        cur = self._first
        while cur.next != None:
            cur = cur.next
        self._last = cur

    def find(self, value):
        cur = self._first
        for i in range(self.ln):
            if cur.value == value:
                return i
            cur = cur.next
        return -1

    def find_min_max(self):
        min = None
        max = None
        if not self.empty():
            min = self._first.value
            max = self._first.value
            cur = self._first
            for i in range(self.ln):
                if cur.value < min:
                    min = cur.value
                if cur.value > max:
                    max = cur.value
                cur = cur.next
        return min, max

    def sort1(self, reverse=False):
        for i in range(self.ln - 1):
            cur = self._first
            for j in range(self.ln - i - 1):
                if j == 0:
                    if (not (cur.value < cur.next.value)) ^ reverse:
                        # cur -> p1 -> p2
                        # p1 -> cur -> p2
                        p1 = cur.next
                        p2 = p1.next

                        p1.next = cur
                        cur.next = p2
                        self._first = p1
                        cur = p1
                else:
                    if (not (cur.next.value < cur.next.next.value)) ^ reverse:
                        # cur -> p1 -> p2 -> p3
                        # cur -> p2 -> p1 -> p3
                        p1 = cur.next
                        p2 = p1.next
                        p3 = p2.next

                        cur.next = p2
                        p2.next = p1
                        p1.next = p3
                    cur = cur.next
        self.set_correct_last()


    def sort2(self, reverse=False):
        for i in range(self.ln - 1):
            cur = self._first
            for j in range(self.ln - i - 1):
                if (cur.value > cur.next.value) ^ reverse:
                    cur.value, cur.next.value = cur.next.value, cur.value

                cur = cur.next



    def change_value(self, old_value, new_value):
        cur = self._first
        for i in range(self.ln):
            if cur.value == old_value:
                cur.value = new_value
            cur = cur.next

    def different_values(self):
        cur1 = self._first
        k = 0
        for i in range(self.ln):
            cur2 = self._first
            ok = True
            for j in range(i):
                if cur2.value == cur1.value:
                    ok = False
                cur2 = cur2.next
            if ok:
                k += 1
            cur1 = cur1.next
        return k

    def reverse(self):
        cur = self._first
        for i in range(self.ln):
            self.push_front(cur.value)
            cur = cur.next
        for i in range(self.ln//2):
            self.pop_back()

    def tolist(self):
        res = []
        cur = self._first
        for i in range(self.ln):
            res.append(cur.value)
            cur = cur.next
        return res



    def pop_identic(self):
        cur1 = self._first
        for i in range(self.ln-1):
            cur2 = self._first
            ok = True
            while cur1 != cur2:
                if cur2.value == cur1.next.value:
                    cur1.next = cur1.next.next
                    self.ln -= 1
                    ok = False
                    break
                cur2 = cur2.next
            if cur2.value == cur1.next.value:
                    cur1.next = cur1.next.next
                    self.ln -= 1
                    ok = False
            if ok:
                cur1 = cur1.next

        self.set_correct_last()


class ListNode2(object):
    def __init__(self, value=None, next=None, prev=None):
        self.value = value
        self.next = next
        self.prev = prev


class List2(object):

    def __init__(self):
        self._first = ListNode2()

    def push_front(self, value):
        if self._first.value is None:
            self._first.value = value
        else:
            self._first = ListNode2(value, self._first)
            self._first.next.prev = self._first

    def push_back(self, value):
        if self._first.value is None:
            self._first.value = value
        else:
            cur = self._first
            while cur.next is not None:
                cur = cur.next
            cur.next = ListNode2(value, None, cur)


    def print(self):
        cur = self._first
        while cur.next is not None:
            print(cur.value, end=' ')
            cur = cur.next
        print(cur.value)

    def clear(self):
        self._first = ListNode2()

    def empty(self):
        if self._first.value is None:
            return True
        return False

    def count(self):
        k = 1
        cur = self._first
        while cur.next is not None:
            k += 1
            cur = cur.next
        return k






    def pop_front(self):
        self._first = self._first.next
        self._first.prev = None

    def pop_back(self):
        cur = self._first
        while cur.next.next is not None:
            cur = cur.next
        cur.next = None

    def pop_value(self, value, every=False):
        if self.count == 1 and self._first.value == value:
            self._first = ListNode()
        cur = self._first
        if self._first.value == value:
            self.pop_front()
        elif every:
            while cur.next is not None:
                if cur.next.value == value:
                    if cur.next is None:
                        cur.next = None
                    else:
                        cur.next = cur.next.next
                    cur.next.prev = cur
                cur = cur.next
            if cur.next.value == value:
                self.pop_back()
        else:
            while cur.next.next is not None:
                if cur.next.value == value:
                    cur.next = cur.next.next
                    cur.next.prev = cur
                    return
                cur = cur.next
            if cur.next.value == value:
                self.pop_back()

    def find(self, value):
        k = 0
        cur = self._first
        while cur is not None:
            if cur.value == value:
                return k
            k += 1
            cur = cur.next
        return -1

    def max(self):
        k = self._first.value
        cur = self._first
        while cur is not None:
            if cur.value > k:
                k = cur.value
            cur = cur.next
        return k

    def min(self):
        k = self._first.value
        cur = self._first
        while cur is not None:
            if cur.value < k:
                k = cur.value
            cur = cur.next
        return k

    def change_value(self, old_value, new_value):
        cur = self._first
        while cur.next is not None:
            if cur.value == old_value:
                cur.value = new_value
            cur = cur.next

    def different_values(self):
        k = 0
        cur = self._first
        while cur.next is not None:
            j = cur.next
            ok = True
            while j.next is not None:
                if cur.value == j.value:
                    ok = False
                j = j.next
            if ok:
                k += 1
            cur = cur.next
        i = self._first
        ok = True
        while i.next is not None:
            if i.value == cur.value:
                ok = False
            i = i.next
        if ok:
            k += 1
        return k

    def pop_identic(self):
        cur = self._first
        while cur.next is not None:
            j = cur.next
            while j is not None:
                if cur.value == j.value:
                    self.pop_value(cur.value)
                j = j.next
            cur = cur.next

    def reverse(self):
        cur = self._first
        prev = None
        while cur.next is not None:
            next = cur.next
            cur.next = prev
            prev = cur
            cur = next
        self._first = prev
        self.push_front(cur.value)
        self._first.prev = None
        cur = self._first
        while cur.next is not None:
            cur.next.prev = cur
            cur = cur.next

    def sort(self, on_value=True):
        j = self._first
        if on_value:
            while j.next is not None:
                cur = self._first
                while cur.next is not None:
                    if cur.value < cur.next.value:
                        cur.value, cur.next.value = cur.next.value, cur.value
                    cur = cur.next
                j = j.next
        else:
            while j.next is not None:
                cur = self._first
                while cur.next is not None:
                    if cur.value < cur.next.value:
                        pup = cur
                        cur = cur.next
                        pup.next = cur.next
                        cur.next = pup
                    cur = cur.next
                j = j.next
        self._first.prev = None
        cur = self._first
        while cur.next is not None:
            cur.next.prev = cur
            cur = cur.next

def create_list():
    list1 = List()
    n = random.randint(1, 10)
    for i in range(n):
        list1.push_back(random.randint(1, 100))
    return list1

print("однонаправленный")
a = create_list()
a.print()

a.push_front(1)
a.print()

a.push_back(1)
a.print()

print(len(a))

print(a.empty())

a.pop_front()
a.print()

a.pop_back()
a.print()

print("---------------------")

b = List()
b.push_back(5)
b.push_back(2)
b.push_back(2)
b.push_back(4)
b.push_back(10)
b.push_back(18)
b.push_back(4)
b.push_back(1)
b.print()

b.pop_all_value(2)
b.print()

b.pop_one_value(1)
b.print()

print(b.find(5))

print(b.find_min_max())

b.change_value(10, 99)
b.print()

print(b.different_values())

b.reverse()
b.print()

b.sort1()
b.print()

b.sort2()
b.print()

b.pop_identic()
b.print()

b.clear()

print("-----------------")
print("двунаправленный")
c = List2()
c.push_back(5)
c.push_back(2)
c.push_back(3)
c.push_back(1)
c.push_back(8)
c.push_back(8)
c.push_back(8)
c.push_back(6)
c.push_front(1)
c.print()

print(c.count())

print(c.empty())

c.pop_front()
c.print()

c.pop_back()
c.print()

c.pop_value(8)
c.print()

print(c.find(1))

print(c.max(), c.min())

c.change_value(2, 10)
c.print()

print(c.different_values())

c.pop_identic()
c.print()

c.reverse()
c.print()

c.sort()
c.print()
```
