# Анализ данных

Создение массивов

```python
file = open("/content/drive/MyDrive/data project.csv")
l = file.readlines()
date_list = []
born_list = []
dead_list = []
growth_list = []
for i in range(1, len(l)):
  l[i] = l[i].split(sep=";")

  date = l[i][0]
  date_list.append(date[3:])

  born = int(l[i][1])
  born_list.append(born)

  dead = int(l[i][2])
  dead_list.append(dead)

  growth = born - dead
  growth_list.append(growth)

file.close()
```

Нахождение максимума и минимума

```python
index_max_growth = growth_list.index(max(growth_list))
index_min_growth = growth_list.index(min(growth_list))

print("Наибольший естественный прирост, человек:")
print(growth_list[index_max_growth], " (", date_list[index_max_growth], ")",
      sep="", end="\n\n")
print("Наименьший естественный прирост, человек:")
print(growth_list[index_min_growth], " (", date_list[index_min_growth], ") ",
      sep="")
```

Построение графика

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 10))
plt.plot(date_list, growth_list, label='прирост населения, человек')
plt.xticks(range(0, len(date_list), 12))
plt.xticks(rotation=90)
plt.xlabel("дата", fontsize=17, fontweight="bold")
plt.ylabel("прирост", fontsize=17, fontweight="bold")
plt.legend(fontsize=17)
```
